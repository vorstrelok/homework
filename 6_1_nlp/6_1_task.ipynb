```python
import pandas as pd
import numpy as np
import sklearn
```

```python
settings = {
    'sep':';',
    'header':None,
    'names':['id','tdate','tname','ttext','ttype','trep', 'trtw', 'tfav','tstcount','tfol','tfrien','listcount'],
    'index_col':'id',
    'parse_dates':[1],
    'date_parser':lambda x: pd.to_datetime(x, unit='s'),
}


positive = pd.read_csv('/home/vorstrel/study/data/nlp_sentiment/positive.csv', **settings)
negative = pd.read_csv('/home/vorstrel/study/data/nlp_sentiment/negative.csv', **settings)
corpus = pd.concat([positive, negative])
```

```python
import sklearn.pipeline
import sklearn.feature_extraction
import sklearn.svm
import sklearn.model_selection
```

```python
tube = sklearn.pipeline.Pipeline([
#     ('vectorize', sklearn.feature_extraction.text.CountVectorizer(max_df=0.9, min_df=5)),
#     ('vectorize', sklearn.feature_extraction.text.TfidfVectorizer(max_df=0.99, min_df=2)),
#     ('vectorize', sklearn.feature_extraction.text.CountVectorizer()),
#     ('vectorize', sklearn.feature_extraction.text.TfidfVectorizer()),
    ('vectorize', None),
    ('classifier', sklearn.svm.LinearSVC())
])
```

```python
grid_search = sklearn.model_selection.GridSearchCV(
    tube,
    param_grid={
        'vectorize':[
            sklearn.feature_extraction.text.CountVectorizer(),
            sklearn.feature_extraction.text.CountVectorizer(max_df=0.95, min_df=2),
            sklearn.feature_extraction.text.TfidfVectorizer(),
            sklearn.feature_extraction.text.TfidfVectorizer(max_df=0.95, min_df=2),
        ]
    },
    n_jobs=-1,
)
```

```python
grid_search.fit(corpus.ttext, corpus.ttype.values)
```
